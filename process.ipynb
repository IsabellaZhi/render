```python
import pandas as pd
```

```python
raw = pd.read_csv("Street_Tree_List-2022-01-30_FILTERED.csv")
raw
```

```python
df = raw[raw['PlantDate'].notna()]
df = df[df['DBH'].notna()]
df = df[df['qCaretaker'].notna()]
df
```

```python
print(df['PlantDate'].dtype)
```

```python
df['PlantDate'] = pd.to_datetime(df['PlantDate'])
today = pd.to_datetime('2024-02-07')
df['Days'] = (today - df['PlantDate']).dt.days
df['Years'] = df['Days']/365
df = df[df['Years'] > 0]

df
```

```python
df.to_csv('filtered_years.csv', index=False)
```
